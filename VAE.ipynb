```python
import tensorflow as tf
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

## Preproccess the images

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255,
                                  horizontal_flip=True,
                                  height_shift_range=.1,
                                  width_shift_range=.1,
                                  brightness_range=(.5,1.5),
                                  zoom_range=[1,1.5])
train_dir = "NewImages/"

train_data = train_datagen.flow_from_directory(
    train_dir,  # Folder containing your image dataset
    target_size=(224,224),  # Resize images to target size
    batch_size=32,
    class_mode=None,  # Use None if you're just augmenting data (no labels)
    shuffle=True
    )

img = next(train_data)
plt.imshow(img[5])
```

## Creating noise for data augmentation

```python
def gaussian_noise(img, mean, stddev, gamma = 1):
    noise=np.zeros(img.shape,np.uint8)
    cv2.randn(noise,mean,stddev)
    noise_img = noise+img
    plt.imshow(noise_img)
    return noise_img
```

```python
gaussian_noise(img[5], 0, .2,6)
```
